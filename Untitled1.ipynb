```python
class jsonParser:

    def parselisting(self, jsonData) -> listingModel:
        listing_model = listingModel()
        listing_model["permalink"] = self.findNaming(jsonData["item"], "nft_id")
        
        listing_model["price"] = float(self.findNaming(jsonData, "base_price")) / 1000000000000000000
        listing_model["symbol"] = self.findNaming(jsonData, "payment_token")["symbol"]
        listing_model["quantity"] = jsonData["quantity"]

        date = datetime.datetime.utcnow()
        utc_time = calendar.timegm(date.utctimetuple())
        listing_model["savingtime"] = utc_time
        listing_model["expirationtime"] = self.getDateTime(self.findNaming(jsonData, 'expiration_date'))

        return listing_model

    def parsebid(self, jsonData) ->bidModel:

        bid_model=bidModel()
        bid_model["permalink"]=self.findNaming(jsonData["item"], "nft_id")
        bid_model["price"]= float(self.findNaming(jsonData, "base_price")) / 1000000000000000000
        bid_model["symbol"]=self.findNaming(jsonData, "payment_token")["symbol"]
        bid_model["quantity"]=jsonData["quantity"]

        date = datetime.datetime.utcnow()
        utc_time = calendar.timegm(date.utctimetuple())
        bid_model["savingtime"]=utc_time
        bid_model["expirationtime"]=self.getDateTime(self.findNaming(jsonData, 'expiration_date'))

        return bid_model

    def findNaming(self, json: json, name: str):
        '''
        please send name with underline_ for words seperator
        :param json:
        :param name:
        :return: json data
        '''
        nameList = [name, name.replace("_", "")]
        positions = self.find(name, "_")
        nameUpper = name
        for position in positions:
            character = nameUpper[position + 1]
            nameUpper = nameUpper.replace("_" + character, "_" + character.upper())
        nameList.append(nameUpper.replace("_", ""))

        for name in nameList:
            try:
                return json[name]
            except:
                pass
        return None


    def find(self, s, ch):
        return [i for i, ltr in enumerate(s) if ltr == ch]


    def getDateTime(self, date):
        try:
            lastTime = datetime.strptime(date, "%Y-%m-%d %H:%M:%S.%f")
            return lastTime
        except Exception as e:
            try:
                lastTime = datetime.strptime(date, "%Y-%m-%d %H:%M:%S")
                return lastTime
            except:
                try:
                    lastTime = datetime.strptime(date, "%Y-%m-%dT%H:%M:%S.%f")
                    return lastTime
                except:
                    try:
                        lastTime = datetime.strptime(date, "%Y-%m-%dT%H:%M:%S")
                        return lastTime
                    except:
                        return date
```
